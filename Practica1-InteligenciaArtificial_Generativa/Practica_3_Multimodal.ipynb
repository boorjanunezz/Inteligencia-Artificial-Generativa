## 3.1 Procesamiento de Imágenes (Vision)
Convertimos imágenes locales a Base64 para su análisis.

```python
import os
import base64
from openai import AzureOpenAI
from dotenv import load_dotenv

load_dotenv()

client = AzureOpenAI(
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    api_key=os.getenv("AZURE_OPENAI_KEY"),
    api_version="2024-08-01-preview" 
)

def encode_image(image_path):
    with open(image_path, "rb") as file:
        return base64.b64encode(file.read()).decode('utf-8')

deployment_name = "gpt-4o"

image_path = "coche.jpg"
if os.path.exists(image_path):
    b64 = encode_image(image_path)
    response = client.chat.completions.create(
        model=deployment_name,
        messages=[{"role": "user", "content": [{"type": "text", "text": "Describe el motor."}, {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{b64}"}}]}]
    )
    print(response.choices[0].message.content)
```

## 3.2 Análisis Estructurado (JSON)
Analizamos una pieza específica (Alternador).

```python
image_path = "pieza_coche.jpg"
if os.path.exists(image_path):
    b64 = encode_image(image_path)
    response = client.chat.completions.create(
        model=deployment_name,
        messages=[
            {"role": "system", "content": "Responde solo en JSON."}, 
            {"role": "user", "content": [{"type": "text", "text": "Estado de la pieza."}, {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{b64}"}}]}
        ],
        response_format={"type": "json_object"}
    )
    print(response.choices[0].message.content)
```
